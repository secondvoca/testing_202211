```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 4, 6, 7])
y = np.array([1, 3, 3, 5, 4])
a = np.dot(x, y) / (x**2).sum()

plt.scatter(x, y)
plt.plot([0, x.max()], [0, a*x.max()])
plt.show()
```

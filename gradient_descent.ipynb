```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f1(x):
    return 3 * x**2 + 6 * x + 1

def f1d(x):
    """f1(x)의 도함수"""
    return 6 * x + 6

xx = np.linspace(-4, 2, 200)

plt.plot(xx, f1(xx), 'k-')
plt.plot(-1, f1(-1), 'ro', markersize=10)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("Object function")
plt.show()
```

```python
step_size = 0.1 # 0.34
```

```python
# start
x = -3

plt.plot(xx, f1(xx), 'k-')
plt.plot(x, f1(x), 'go', markersize=10)
plt.text(x + 0.2, f1(x) + 0.1, "start")
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'b--')
plt.ylim(-5, 15)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("optimization by using steepest gradient descent")
plt.grid(True)
plt.show()

print("start: x_0 = {:.2f}, g_0 = {:.2f}".format(x, f1d(x)))
```

```python
# k = 1
x = x - step_size * f1d(x)

plt.plot(xx, f1(xx), 'k-')
plt.plot(x, f1(x), 'go', markersize=10)
plt.text(x + 0.2, f1(x) + 0.1, "1st step")
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'b--')
plt.ylim(-5, 15)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("optimization by using steepest gradient descent")
plt.grid(True)
plt.show()

print("1st step: x_1 = {:.2f}, g_1 = {:.2f}".format(x, f1d(x)))
```

```python
# k = 2
x = x - step_size * f1d(x)

plt.plot(xx, f1(xx), 'k-')
plt.plot(x, f1(x), 'go', markersize=10)
plt.text(x + 0.2, f1(x) + 0.1, "2nd step")
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'b--')
plt.ylim(-5, 15)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("optimization by using steepest gradient descent")
plt.grid(True)
plt.show()

print("2nd step: x_1 = {:.2f}, g_1 = {:.2f}".format(x, f1d(x)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline

def f1(x):
    return 3 * x**2 + 6 * x + 1

def f1d(x):
    """f1(x)의 도함수"""
    return 6 * x + 6

xx = np.linspace(-4, 2, 200)

plt.plot(xx, f1(xx), 'k-')

step_size = 0.1 # 0.34

# start
x = -3
plt.plot(x, f1(x), 'go', markersize=10)
plt.text(x + 0.2, f1(x) + 0.1, "start")
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'b--')
print("start: x_0 = {:.2f}, g_0 = {:.2f}".format(x, f1d(x)))

# k = 1
x = x - step_size * f1d(x)
plt.plot(x, f1(x), 'go', markersize=10)
plt.text(x + 0.2, f1(x) + 0.1, "1st step")
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'b--')
print("1st step: x_1 = {:.2f}, g_1 = {:.2f}".format(x, f1d(x)))

# k = 2
x = x - step_size * f1d(x)
plt.plot(x, f1(x), 'go', markersize=10)
plt.text(x + 0.2, f1(x) + 0.1, "2nd step")
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'b--')
print("2nd step: x_2 = {:.2f}, g_2 = {:.2f}".format(x, f1d(x)))


plt.ylim(-5, 15)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("optimization by using steepest gradient descent")
plt.grid(True)
plt.show()
```
